```python
from collections import defaultdict
from itertools import combinations
from datetime import datetime

class Apriori:
    def __init__(self, filename):
        self.file = filename
        self.transection = list() # set of item
        self.C_itemset = defaultdict(int) # k-item set for the item in dict format
        self.GlobalFreqItem = {}
        self.candidates=defaultdict(int) 
        
        
        
        

    def LoadData(self):
        with open(self.file,'r') as file:
            data = file.readlines()
            for i in data:
                temp = frozenset(map(int,i.split()))
                self.transection.append(temp) 
            
                for item in temp:
                    self.C_itemset[frozenset([item])] +=1
   
    def Count_Support(self,candidates, k):
        C_support = defaultdict(int)
        for i in self.transection:
            if len(i) == k:
                if i in candidates:
                    C_support[i]+=1
            
            elif len(i)>k:
                subset = list(combinations(i,k))
                for j in subset:
                    
                    if frozenset(j)in candidates:
                            C_support[frozenset(j)]+=1
        return C_support
    
    def freq_item_dict(self, item_dict, support):
        
        return {k:v for k,v in item_dict.items() if v >= support}
        
    
     
    
    def apriori(self, minFreq):
        time_start = datetime.now()
        self.LoadData()
    
        minSup = minFreq * len(self.transection)
 
        Lk = self.freq_item_dict(self.C_itemset, minSup)
        K=1
        self.GlobalFreqItem[K] = Lk
        while len(Lk) > 0:
            K+=1
            self.candidates = set([i.union(j) for i in Lk for j in Lk if len(i.union(j)) == K])
    
            C_k = self.Count_Support(self.candidates, K )
            Lk = self.freq_item_dict(C_k, minSup)
            self.GlobalFreqItem[K] = Lk
           
            
        
        time = datetime.now() - time_start
    
        return self.GlobalFreqItem, time
```

```python

for i in [0.0001,0.0002,0.0003,0.0004,0.0005]:
    
    model = Apriori('trans.txt')
    Items,t = model.apriori(minFreq=i)
    print("Min Support:",i )
    for j in Items.keys():
        print("lv",j)
        print("no.of freq element",len(Items[j]))
        #print(,'\n',Items[j])
        print()
    print("Time using:",t)
    print()
    print()
```
